## <u>WiFi Coverage Problem</u>

Shopee has decided to get master WiFi network hubs from two different companies namely GeoFi and AirFi.
However, there is a big problem, these two networking device manufacturing companies have a long history of rivalry. 
They developed their network technology in such a way that these two networks cancel each other. Meaning, if an engineer is inside the network range of both GeoFi and AirFi network hubs, he/she will not get any signal from either of these two network hubs (see picture).



In the picture above, the two points (a,b) in the intersected area do not get any WiFi signal. Also the three points (c,d,e) outside the area of the circles do not have any WiFi signal.

Installing networking devices without complex calculation might bring about networking
blackout at different locations on the office floor, hence, Shopee IT decided to install only one GeoFi network hub on the office floor. The same rule goes for the AirFi network hub. Each hub has a range R and they cover a circular area of radius R centering the position of the network hub. Shopee IT hopes that every engineer is able to receive the WiFi signal from one of the two new network hubs. 

Shopee IT wants your answer for the following:

**Given coordinates of each engineer, coordinates of the GeoFi network hub (Cg) and AirFi network hub (Ca) and the range of these network hubs Rg(GeoFi) and Ra(AirFi), find the number of engineers that will not get any WiFi signal.**

### <u>Input</u>

Only one test case. The test case begins with an integer **N (1  N  105)**, the number of engineers in the office.
Each of the next N lines will have two integers **x, y (0  x, y  108)** representing the coordinates of the engineers.
The next line contains four integers **Xg, Yg representing Cg, and Xa, Ya representing Ca**.

The next line contains an integer number **Q (1  Q  105)**, representing the number of queries.
Each of the next Q lines will have two integers **Rg, Ra (0  Rg, Ra  108)**.

### <u>Output</u>
For each query (given value of Rg, Ra), print the number of engineers that will not get any WiFi signal.

```python
import math
```

```python
nb_engineers = int(input("Number of engineers in office: "))
eng_coor = [] # list of engineers' coordinates
for n in range(nb_engineers):
    x, y = map(int,input("Coordinates of engineer " + str(n+1) + ": ").split())
    eng_coor.append((x,y))
```

```python
Xg, Yg = map(int, input("Coordinates of GeoFi network hub's centre: ").split())
Cg = (Xg, Yg)

Xa, Ya = map(int, input("Coordinates of AirFi network hub's centre: ").split())
Ca = (Xa, Ya)
```

```python
nb_queries = int(input("Number of queries: "))
query_queue = [] # list of radius queries
for q in range(nb_queries):
    Rg, Ra = map(int, input("Rg Ra of Query " + str(q+1) + ": ").split())
    query_queue.append((Rg, Ra))
```

**How do we know if an engineer, i, is in one circle, c? If the length between (x,y) and centre of c (Xc, Yc) < radius of c, Rc.**

**Then, if the above holds true for two circles, the engineer must be in their intersection**.

```python
def distance(coor1, coor2):
    """
    This function calculates the distance between two coordinate points
    """
    d = round(math.sqrt((coor1[0] - coor2[0])**2 + (coor1[1] - coor2[1])**2),2)
    return d
```

```python
for q in query_queue:
    noWifi = []
    for engineer in eng_coor:
        if ((distance(Cg, engineer) <= q[0]) and (distance(Ca, engineer) <= q[1])) or ((distance(Cg, engineer) > q[0]) and (distance(Ca, engineer) > q[1])):
            noWifi.append(engineer)
    print(len(noWifi))
```
